## 1) Python Data Structures

### Tuples

```python

# Tuples are immutable and support unpacking

point = (12.5, 7.0)
x, y = point
print("x:", x, "y:", y)

def arith(a,b):

  return a*b,a+b,a-b,[b,a]

mult,sum,subt,rev = arith(x,y)

print("x*y:", mult, "x+y:", sum,"x-y",subt,"rev",rev)
```

### Sets

```python

# Sets keep unique items; order is not guaranteed
a = {1,2,3,3,2}
b = {3,4,5}
print("a:", a)
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference a-b:", a - b)

c = set()

c.add(7)
c.add(8)
print(c)
c.remove(7)
print(c)

if 5 in b:
  print("5 is in b")
else:
  print("5 is not in b")
```

## 2) Python OOP — Classes & Objects

```python
class Counter:
    def __init__(self, start=0):
        self.value = int(start)
    def inc(self, step=1):
        self.value += step
    def dec(self,step=1):
        self.value -= step
    def reset(self):
        self.value = 0
    def __repr__(self):
        return f"Counter(value={self.value})"

c = Counter(5)
c.inc()
print(c)
c.inc(3)
print(c)
c.dec()
print(c)
c.dec(2)
print(c)
```

### Try…Except

```python

def safe_div(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return float("inf")  # convention for division by zero
    except TypeError as e:
        print("Type error:", e)
        return None

print(safe_div(10,2))
print(safe_div(1,0))
print(safe_div("x",5))
```

```python
import os

filename = "notes.txt"
target_string = "425"

# ==========================================
# 1. Setup: Recreate your exact notes.txt file
# ==========================================
# Adding the header text
file_lines = [
    "EWU CSE", 
    "Lab 2: File I/O", 
    "Hello, world!", 
    ""  # Empty line based on your prompt
]

# Adding the numbers from 5 to 495 (counting by 5)
for i in range(5, 500, 5):
    file_lines.append(str(i))

# Writing to the file
with open(filename, "w") as file:
    for line in file_lines:
        file.write(f"{line}\n")

# ==========================================
# 2. File Handling
# ==========================================
def read_file_to_list(filepath):
    """Reads a file and returns a list of strings, removing newlines."""
    data = []
    with open(filepath, "r") as f:
        for line in f:
            data.append(line.strip()) 
    return data

# ==========================================
# 3. Simple (Linear) Search
# ==========================================
def linear_search(data_list, target):
    """Searches item by item. Returns (index, iterations)."""
    iterations = 0
    for index, item in enumerate(data_list):
        iterations += 1
        if item == target:
            return index, iterations
            
    return -1, iterations

# ==========================================
# 4. Binary Search
# ==========================================
def binary_search(data_list, target):
    """Searches by dividing the list in half. Returns (index, iterations)."""
    iterations = 0
    low = 0
    high = len(data_list) - 1
    
    while low <= high:
        iterations += 1
        mid = (low + high) // 2
        guess = data_list[mid]
        
        if guess == target:
            return mid, iterations
        elif guess > target:
            high = mid - 1  
        else:
            low = mid + 1   
            
    return -1, iterations 

# ==========================================
# 5. Execution and Output
# ==========================================
if __name__ == "__main__":
    
    # Read data from the file
    file_data = read_file_to_list(filename)
    print(f"Target to find: '{target_string}'\n")

    # Run Linear Search
    lin_index, lin_iters = linear_search(file_data, target_string)
    print(f"--- Simple (Linear) Search ---")
    if lin_index != -1:
        print(f"Found '{target_string}' at original index {lin_index}.")
    else:
        print(f"'{target_string}' not found.")
    print(f"Loop iterations required: {lin_iters}\n")

    # Run Binary Search
    # We must sort the list first! (Note: it sorts alphabetically as strings)
    sorted_data = sorted(file_data) 
    
    bin_index, bin_iters = binary_search(sorted_data, target_string)
    print(f"--- Binary Search ---")
    print(f"(Note: Data was sorted alphabetically for binary search)")
    if bin_index != -1:
        print(f"Found '{target_string}' at sorted index {bin_index}.")
    else:
        print(f"'{target_string}' not found.")
    print(f"Loop iterations required: {bin_iters}")

    # Clean up the file (optional, remove if you want to keep notes.txt)
    if os.path.exists(filename):
        os.remove(filename)
```
