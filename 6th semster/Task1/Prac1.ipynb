```python
file_path = "try.txt"

with open(file_path, "w", encoding="utf-8") as f:
    data = []
    for i in range(2, 300, 3):
        f.write(f"{str(i)}\n")
```

```python
target = int(input("Enter a number from 2 - 300, to find it's index\n> "))
flag = 0
index = -1
with open(file_path, "r", encoding="utf-8") as f:
    data = f.readlines()
    for i in data:
        index += 1
        if target == int(i):
            flag=1
            break


if flag == 1:
    print(f"{target} is found at > {index} index/line")

else:
    print("Sorry")
```
